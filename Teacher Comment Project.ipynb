```python
import csv

#All sub functions in this cell

def firstparagraph(name,pronoun1,reflection): 
    '''Function outputs the introductory standardized comment paragraph along with 
    student's personal statement'''
    
    return f"In the first semester, we covered the fundamentals of Python, "\
           "including loops, conditionals, functions, and dictionaries. "\
           "In addition, we explored connections between math and "\
           "computer science, including plotting, simulating the game of craps, "\
           f"and optimizing customer service at a bank.\n\nIn {name.strip()}'s "\
           f"self reflection, {pronoun1} wrote: \n\n{reflection.strip()}"


def quiz_avg(quiz1,quiz2,quiz3,quiz4):
    '''Function outputs quiz average'''
    
    return (quiz1+quiz2+quiz3+quiz4)/4

def ps_avg(ps1,ps2,ps3,ps4,ps5,ps6):
    '''Function outputs problem set average'''
    
    return (ps1+ps2+ps3+ps4+ps5+ps6)/6 

def grade(ps_avg,quiz_avg,exam,bank):
    '''Function outputs final letter grade'''
    
    grade = (ps_avg+quiz_avg+exam+bank)/4
    if grade >= 93:
        return 'A'
    elif grade >= 90:
        return 'A-'
    elif grade >= 86:
        return 'B+'
    elif grade >= 83:
        return 'B'
    elif grade >= 80:
        return 'B-'
    elif grade >= 76:
        return 'C+'
    elif grade >= 73:
        return 'C'
    elif grade >= 69:
        return 'C-'
    elif grade >= 65:
        return 'D'
    

def ps_sentence(ps,pronoun1,pronoun3):
    '''Function outputs sentence about user ps depending on grade'''
    
    if ps >= 93:
        return  f"\n\n{pronoun3.title()} problem set average was {ps:.2f} "\
                f"which indicates that {pronoun1} had a solid understanding of "\
                f"the majority of the topics covered this semester and that {pronoun1} "\
                "was consistent about meeting deadlines."
    if 80 <= ps <92:
        return  f"\n\n{pronoun3.title()} problem set average was "\
                f"{ps:.2f} which indicates that {pronoun1} understood "\
                "most of the topics covered this semester and that "\
                f"{pronoun1} met the deadline most of the time."        
    else:           
        return  f"\n\n{pronoun3.title()} problem set average was "\
                f"{ps:.2f} which indicates that {pronoun1} had gaps in "\
                f"{pronoun3} understanding and was inconsistent about meeting deadlines."
        

def quiz_sentence(quiz_avg,final_exam,pronoun1,pronoun3):
    '''Function outputs sentence about user quiz and final exam 
    depending on grade'''
    
    final_avg = (quiz_avg+final_exam)/2
    if final_avg >= 93: 
        return  f" {pronoun3.title()} quiz average and final exam score "\
                f"were {quiz_avg} and {final_exam} respectively, which "\
                f"demonstrated that {pronoun1} grasped the new content "\
                "thoroughly and could hand write algorithms without using outside resources."
    if 80 <= final_avg < 92:
        return f" {pronoun3.title()} quiz average and final exam score "\
               f"were {quiz_avg} and {final_exam} respectively, which "\
               f"demonstrated that {pronoun1} understood most of the new "\
               "content, but still need some improvement on hand writing "\
               "algorithms without using outside resources."
    else:
        return  f" {pronoun3.title()} quiz average and final exam "\
                f"score were {quiz_avg} and {final_exam} respectively, "\
                f"which demonstrated that {pronoun1} did not grasp the "\
                "new content thoroughly and could not hand write "\
                "algorithms without using outside resources."
        

def bank_paragraph(bankgrade, hackathon, pronoun1, pronoun2, pronoun3,pronoun4):
    '''Outputs paragraph that explains hackathon and bank project
    performance depending on bankgrade and hackathon participation'''
    
    if bankgrade >= 90:
        collaboration = "good"
    elif bankgrade >= 80:
        collaboration = "mediocre"
    else:
        collaboration = "unsatisfactory"        
       
    if hackathon == 'Yes':
        extra = f"In addition, {pronoun1} demonstrated significant "\
        "enthusiasm and dedication when we participated in the Hackathons"
        effort = f"{pronoun1} chose to use these opportunities to "\
        f"seriously challenge {pronoun4}, think creatively, and advance {pronoun3} skills."
    
    if hackathon == "No":
        extra = f"In addition, {pronoun1} tended to lose focus when we participated in the Hackathons"
        effort = f"{pronoun1} often lost focus instead of working on advancing {pronoun3} skills."
    
    return f"\n\n{pronoun3.strip().title()} group Bank Project grade was an "\
           f"{bankgrade}, which indicated that {pronoun3} effort, time management, "\
           f"and collaboration relating to larger group assignments was "\
           f"{collaboration}. {extra}. These assignments allowed students "\
           f"to put forth as much effort as they wished to, and {effort}"


def daily_activity(team_worker, extra_help, class_grade,pronoun1,pronoun2,pronoun3,pronoun4):
    '''Function outputs paragraph explaining more detailed information about student â€” 
    attributes as a team worker & willingness to ask questions'''
    
    if class_grade == 'A' or class_grade == 'A-' or class_grade == 'B+':
        if team_worker == 'No':
            teamwork = f"On the daily coding activities, {pronoun1} worked "\
            f"mostly by {pronoun4} since the material came easily to {pronoun2}, "\
            f"but {pronoun1} was happy to help a classmate when {pronoun1} got stuck."
        else:
            teamwork = f"On the daily coding activities, {pronoun1} "\
            "worked with others a fair amount. This collaboration seems to help "\
            f"{pronoun2} and help guide {pronoun2} thinking."
        if extra_help == "Yes":
            extrahelp= f"{pronoun1} wasn't shy at all and it proved to help {pronoun2} tremendously."
        else:
            extrahelp = f"{pronoun1} rarely if ever needed to do so "\
            f"because {pronoun1} was so easily able to pick up the concepts in class."
        
        stack = f"{pronoun1} wisely utilized these resources to troubleshoot "\
        f"and gain inspiration and internalize the solutions that {pronoun1} found."
        
    else:
        if team_worker == "No":
            teamwork = f"On the daily coding activities, it would have been "\
            f"beneficial for {pronoun2} to work with {pronoun3} classmates rather than working alone."
        else:
            teamwork = f"On the daily coding activities, {pronoun1} should have "\
            f"tried to attempt problems by {pronoun4} more often. Working with others "\
            f"extensively has gotten in the way of {pronoun3} own personal learning sometimes."
        if extra_help == "Yes":
            extrahelp = f"{pronoun1} asked questions frequently. However, "\
            f"{pronoun1} would too often ask questions to get the answer "\
            f"rather than internalize the concept. This led to {pronoun2} not "\
            "truly learning from asking pointed questions."
        else:
            extrahelp = f"{pronoun1} rarely if ever did so, which I think "\
            f"would have really helped {pronoun2} to solidify {pronoun3} skills."
        
        stack = f"{pronoun1} sometimes used these resources without fully understanding what the code did."
    
    return f"\n\n{teamwork} In regard to using available resources like StackOverflow, "\
            f"{stack} In terms of asking for extra help, {extrahelp}"
           

def lastmsg(grade,name,pronoun1,pronoun3):
    '''Function that outputs final personalized message to particular student
    depending on the final grade'''
    
    if grade == 'A' or grade == 'A-':
        return f"\n\n{name.strip()}'s grade for the semester is a {grade.strip()}. "\
               f"I look forward to {pronoun3} continued success next semester as we begin to "\
               "explore more advanced computer science and data science topics."
    
    elif grade == 'B+' or grade == 'B':
        return f"\n\n{name.strip()}'s grade for the semester is a {grade.strip()}. "\
               f"While {name} has managed to perform decently in this course, I know "\
               f"{pronoun1} has the potential to master computer science concepts this coming semester."
    else:
        return f"\n\n{name.strip()}'s grade for the semester is a {grade}. "\
               f"If {name.strip()} can commit to putting forth consistent effort "\
               "into the daily and nightly work and to stopping by for extra help regularly, "\
               f"then I am confident that {pronoun1} will enjoy more success in the spring semester."
```

```python
#Primary Function


def main():
    '''Main function that carries out file writing'''
    
    with open('teacher.csv') as f:
        data = csv.reader(f, delimiter = ",")
        header = f.readline().split(',')
        for row in data:       #each row correlates to each student
            gender = row[1]
        
            #Assigning gender pronouns
            if gender == "female":          
                pronoun1,pronoun2,pronoun3,pronoun4 = 'she','her','her','herself'       
            elif gender == 'male':
                pronoun1, pronoun2,pronoun3,pronoun4 = 'he', 'him','his','himself'   
            else:
                pronoun1,pronoun2,pronoun3,pronoun4 = 'they','them','their','themself'
             
            #Defining all variables prior to use in terms of position in csv file row
            name = row[0].strip()
            reflection=row[-1]
            ps = ps_avg(int(row[2]),int(row[3]),int(row[4]),int(row[5]),int(row[6]),int(row[7]))
            quiz= quiz_avg(int(row[8]),int(row[9]),int(row[10]),int(row[11]))
            final_exam = int(row[12])
            bank = int(row[13])
            hackathon = row[-4]
            team_worker = row[-3]
            extra_help = row[-2]
            class_grade = grade(ps,quiz,final_exam,bank)
            
            #Create new student file each time
            comment = open(f"{name}.txt","w")
            
            #Writing file content using previously defined functions
            comment.write(firstparagraph(name,pronoun1,reflection))
            comment.write(ps_sentence(ps,pronoun1,pronoun3))
            comment.write(quiz_sentence(quiz,final_exam,pronoun1,pronoun3))
            comment.write(bank_paragraph(bank, hackathon,pronoun1, pronoun2, pronoun3,pronoun4))
            comment.write(daily_activity(team_worker, extra_help, class_grade,pronoun1, \
            pronoun2,pronoun3,pronoun4))
            comment.write(lastmsg(class_grade,name,pronoun1,pronoun3))
            
            #closing file to allow program to finish writing content properly
            comment.close()
```

```python
#Running main function to carry out algorithm

if __name__ == '__main__':
    main()
```
